```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('startup_funding.csv')

data.dropna(subset = ['IndustryVertical'], inplace = True)
data['AmountInUSD'].fillna(0,inplace=True)

data['IndustryVertical'].replace('eCommerce','Ecommerce',inplace=True)
data['IndustryVertical'].replace('ECommerce','Ecommerce',inplace=True)
data["AmountInUSD"]=data["AmountInUSD"].apply(lambda x: float(str(x).replace(",","")))

company = []
fund = []

for i in data['IndustryVertical']:
    company.append(i)
for i in data['AmountInUSD']:
    fund.append(i)
    
cmpny = np.array(company,dtype = 'str')
fnd = np.array(fund,dtype='int64')

dic={}
for i in range(len(cmpny)):
    if dic.get(cmpny[i], 0) == 0:
        dic[cmpny[i]] = fnd[i]
    else:
        dic[cmpny[i]] += fnd[i]
s=0
for item in dic.items():
    s += item[1]

for item in dic.items():
    print(item[0], format(item[1] * 100 / s, '.2f'))
```

```python
import pandas as pd 
import numpy as np 
df_start=pd.read_csv('startup_funding.csv',encoding='utf-8')
df_start['IndustryVertical'].replace('eCommerce','Ecommerce',inplace=True)
df_start['IndustryVertical'].replace('ECommerce','Ecommerce',inplace=True)
df_start['IndustryVertical'].replace('ecommerce','Ecommerce',inplace=True)
## Converting "AmountInUSD" into numeric format 
df_start["AmountInUSD"] = df_start["AmountInUSD"].apply(lambda x: float(str(x).replace(",",""))) 
df_start["AmountInUSD"] = pd.to_numeric(df_start["AmountInUSD"]) 
industry_amount=df_start.groupby('IndustryVertical')['AmountInUSD'].sum().sort_values(ascending=False)[0:5]
industry=industry_amount.index 
amountIndustry=industry_amount.values 
perIndustry=np.true_divide(amountIndustry, amountIndustry.sum())*100
for i in range(len(industry)):
    print(industry[i],format(perIndustry[i],'.2f'))
```
